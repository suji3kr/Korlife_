```python
cloud=open('d:/data/wordcloud_ex.txt', encoding='UTF-8').read()
cloud
```

```python
import re #정규화 표현식 사용 모듈
cloud=re.sub('[^가-힣]',' ',cloud) #한글이 아니면(^) 공백처리
cloud
```

```python
#hannanum 만들기 -> 명사추출
#! pip install konlpy
import konlpy
hannanum=konlpy.tag.Hannanum()
```

```python
#명사추출
nouns = hannanum.nouns(cloud)
nouns
```

```python
#데이터 프레임으로 변환
import pandas as pd
df_word=pd.DataFrame({'word':nouns})
df_word
```

```python
#글자 수 추가
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
#두 글자 이상 단어만 남김
df_word = df_word.query('count>=2')
df_word.sort_values('count')
```

```python
#단어 빈도 구하기:단어별분리->빈도구하기->내림차순 정렬
df_word= df_word.groupby('word',as_index=False)\
                .agg(n=('word','count'))\
                .sort_values('n',ascending=False)
df_word
```

```python
#단어 빈도 상위 20개 추출
top20=df_word.head(20)
top20
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
 
plt.rcParams.update({'font.family':'Malgun Gothic',
                     'figure.dpi':'120',
                     'figure.figsize':[6.5, 6]})
```

```python
#막대 그래프 
sns.barplot(data=top20, y='word', x='n')
```

```python
#데이터 프레임을 딕셔너리로 변환
dic_word = df_word.set_index('word').to_dict()['n']
dic_word 
```

```python
#! pip install wordcloud
```

```python
font='malgun.ttf'
```

```python
#Word cloud 만들기
from wordcloud import WordCloud
wc=WordCloud(random_state=1234,
            font_path=font,
             width=400,
              height=400,
               background_color='white')
```

```python
#워드 클라우드 출력하기
img_wordcloud = wc.generate_from_frequencies(dic_word)
plt.figure(figsize=(10, 10))
plt.axis('off')
plt.imshow(img_wordcloud)
```

```python
#mask 적용
import PIL  #(Python Imaging Library --현재는 Pillow Library가 사실상 표준)
import PIL.Image
icon= PIL.Image.open('d:/data/cloud.png')
```

```python
#png 파일로 mask 생성
import numpy as np  #***
img = PIL.Image.new('RGB', icon.size,(255,255,255))
img.paste(icon, icon)
img=np.array(img)
```

```python
#WordCloud만들기
wc=WordCloud(random_state=1234, #난수고정
             font_path=font,    #폰트설정
             width=400,         #가로 크기
             height=400,        #세로 크기
             background_color='white',  #배경색
             mask = img)        #마스크 설정
```

```python
img_wordcloud=wc.generate_from_frequencies(dic_word)

# 워드 클라우드 출력
plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(img_wordcloud)
```

```python
wc=WordCloud(random_state=1234, 
             font_path=font,
             width=400,
             height=400,
             background_color='white',
             mask=img,
             colormap='prism')
```

```python
img_wordcloud=wc.generate_from_frequencies(dic_word)

#워드 클라우드 출력
plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(img_wordcloud)
```
